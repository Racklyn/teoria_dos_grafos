```python
from math import inf

class Grafo:

    # Grafo ponderado não-direcionado, sem arestas paralelas

    n_vertices = 0
    matriz = []

    def __init__(self, n_vertices):
        self.n_vertices = n_vertices
        self.matriz = [ [ 0 for i in range(n_vertices) ] for j in range(n_vertices) ]

    def __str__(self):
        result = ''
        for l in self.matriz:
            result += ' '.join([str(i) for i in l]) + '\n'
        return result

    def inserir_aresta(self,vi,vj,p):
        self.matriz[vi][vj] = p
        self.matriz[vj][vi] = p

    def remover_aresta(self,vi,vj):
        self.matriz[vi][vj] = 0
        self.matriz[vj][vi] = 0

    def existe_aresta(self,vi,vj):
        return self.matriz[vi][vj] > 0


    def getGrauVertice(self, vertice):
        grau = 0
        for i in range(self.n_vertice):
            if self.matriz[vertice][i] > 0:
                grau += 1
        return grau

    def getGrauGrafo(self):
        grau = 0
        for v in range(self.n_vertices):
            grau += self.getGrauVertice(v)
        return grau

    def existeLoop(self):
        for i in range(self.n_vertices):
            if (self.matriz[i][i]>0): return True

        return False
    

    def dijkstra(self, inicio):
        # informações sobre os vértices:
        v_info = [{'dist': inf, 'pai': None, 'visitado': False} for _ in range(self.n_vertices)]
        v_info[inicio]['dist'] = 0
        

        while True:
            menor_dist = inf
            v_atual = -1
            for i in range(self.n_vertices):
                if v_info[i]['dist'] < menor_dist and not v_info[i]['visitado']:
                    v_atual = i
                    menor_dist = v_info[i]['dist']

            if v_atual == -1: break

            v_info[v_atual]['visitado'] = True
            
            # Atualizar dist de filhos de v_atual:

            for i_filho in range(self.n_vertices):
                p = self.matriz[v_atual][i_filho]
                if p > 0:
                    if v_info[v_atual]['dist'] + p < v_info[i_filho]['dist']:
                        v_info[i_filho]['dist'] = v_info[v_atual]['dist'] + p
                        v_info[i_filho]['pai'] = v_atual



        # Print do resultado:
        for i, v in enumerate(v_info):
            print(f'{i} - {v}')


        
```

```python
g = Grafo(5)

g.inserir_aresta(0,1,4)
g.inserir_aresta(0,2,2)
g.inserir_aresta(1,2,1)
g.inserir_aresta(1,3,1)
g.inserir_aresta(1,4,3)
g.inserir_aresta(2,4,6)
g.inserir_aresta(3,4,1)

print(g)

g.dijkstra(0)
```
