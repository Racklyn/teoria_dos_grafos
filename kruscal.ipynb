```python
class Grafo:
    # Grafo ponderado não-direcionado, sem arestas paralelas

    n_vertices = 0
    matriz = []

    def __init__(self, n_vertices):
        self.n_vertices = n_vertices
        self.matriz = [ [ 0 for i in range(n_vertices) ] for j in range(n_vertices) ]

    def __str__(self):
        result = ''
        for l in self.matriz:
            result += ' '.join([str(i) for i in l]) + '\n'
        return result

    def inserir_aresta(self,vi,vj,p):
        self.matriz[vi][vj] = p
        self.matriz[vj][vi] = p

    def remover_aresta(self,vi,vj):
        self.matriz[vi][vj] = 0
        self.matriz[vj][vi] = 0

    def existe_aresta(self,vi,vj):
        return self.matriz[vi][vj] > 0


    def getGrauVertice(self, vertice):
        grau = 0
        for i in range(self.n_vertice):
            if self.matriz[vertice][i] > 0:
                grau += 1
        return grau

    def getGrauGrafo(self):
        grau = 0
        for v in range(self.n_vertices):
            grau += self.getGrauVertice(v)
        return grau

    def existeLoop(self):
        for i in range(self.n_vertices):
            if (self.matriz[i][i]>0): return True

        return False
    


    def kruskal(self):

        grupos = [[i] for i in range(self.n_vertices)]
        arestas = [] # lista com arestas ordenadas por peso. Formato: (v1, v2, p)
        arestasRestantes = []

        # criar lista com aresta por ordem de peso (crescente)
        for i in range(self.n_vertices):
            for j in range(i, self.n_vertices):
                p = self.matriz[i][j]
                if p > 0:
                    arestas.append((i, j, p))

        arestas.sort(key=lambda x:x[2]) # ordenando por peso


        # loop por lista de arestas ordenadas
        for a in arestas:
            if len(grupos) == 1: break
            # pegar vértices A e B da aresta em questão
            v1, v2, p = a
            
            # se A e B estão no mesmo grupo (mesma lista dentro de 'grupos'):
            auxGrupos = grupos.copy()
            g_v1 = -1
            g_v2 = -2

            for i in range(len(grupos)):
                g = grupos[i]

                if v1 in g:
                    g_v1 = i

                if v2 in g:
                    g_v2 = i
                
                if g_v1 == g_v2:
                    break

            if g_v1 != g_v2:
            # juntar grupo de A com grupo de B em uma lista só
                grupos[g_v1] = auxGrupos[g_v1] + auxGrupos[g_v2]
                grupos.pop(g_v2)

                arestasRestantes.append(a)

                print(f'Grupos: {grupos}')



        
        print("-"*20)
        for a in arestasRestantes:
            print(f'V1: {a[0]}, V2: {a[1]}, Peso: {a[2]}')

```
